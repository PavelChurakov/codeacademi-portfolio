# U.S. Medical Insurance Costs

Hello everyone!
In this notebook, we analyze health insurance costs. Our analysis is based on data from the Insurances.csv file
The Insurances.csv file contains data on 1338 people. He has seven fields [age,sex,bmi,children,smoker,region,charges]. We are trying to understand how the cost of insurance depends on age, gender, BMI, children, smokers, regions. For a better understanding, we create several dashboards and visualizations.
We will find max, min, average, median and other indicators.
Let's go to!

1)  Import data from insurance.csv

```python
import csv
import matplotlib.pyplot as plt

csvfile = open("insurance.csv", 'r',newline='')
dict_reader = csv.DictReader(csvfile)
lst_dict_reader = list(dict_reader)
fieldnames = lst_dict_reader[0].keys()
```

Show first 10 rows

```python
for row in lst_dict_reader[:10]:
    print(row)
```

Show field names

```python
print(fieldnames)
```

2)Let's start looking for some key features of our data set.

```python
#find avarege age (avarege_age)
list_age = [int(row["age"]) for row in lst_dict_reader]
average_age = sum(list_age)/len(list_age)
print("Avarege age: {} years.".format(round(average_age)))
```

```python
#find avarege BMI (avarege_bmi)
list_bmi = [float(row["bmi"]) for row in lst_dict_reader]
average_bmi = sum(list_bmi)/len(list_bmi)
print("Avarege BMI: {}.".format(round(average_bmi,2)))
```

```python
#find avarege charge (avarege_charge)
list_charges = [float(row["charges"]) for row in lst_dict_reader]
average_charge = sum(list_charges)/len(list_charges)
print("Avarege charge: {}.".format(round(average_charge,2)))
```

```python
#find max charge (max_charge)
max_charge = 0
dict_max = {}
for row in lst_dict_reader:
    if float(row['charges'])> max_charge:
        max_charge = float(row['charges'])
        dict_max = row
print(dict_max)
print("The max charge: {}".format(max_charge))
        
```

```python
#find min charge (min_charge)
min_charge = max_charge
dict_min = {}
for row in lst_dict_reader:
    if float(row['charges'])< min_charge:
        min_charge = float(row['charges'])
        dict_min = row
print(dict_min)
print("The min charge: {}".format(min_charge))
```

```python
#find the distribution by region and show gistogram
list_region = [row["region"] for row in lst_dict_reader]
dict_count_different_region = {}
for region in list_region:
    if region in dict_count_different_region:
        dict_count_different_region[region] += 1
    else:
        dict_count_different_region[region] = 1
print(dict_count_different_region)
plt.xlabel("Regions")
plt.ylabel("Number of people")
plt.bar(dict_count_different_region.keys(),dict_count_different_region.values())
plt.show()
```

```python
#find the distribution by smoker
list_smoker = [row["smoker"] for row in lst_dict_reader]
dict_count_smoker = {}
for smok in list_smoker:
    if smok in dict_count_smoker:
        dict_count_smoker[smok] += 1
    else:
        dict_count_smoker[smok] = 1
print(dict_count_smoker)
plt.xlabel("Smok")
plt.ylabel("Number of people")
plt.bar(dict_count_smoker.keys(),dict_count_smoker.values())
plt.show()
```

```python
#find the distribution by number of children
list_children = [int(row['children']) for row in lst_dict_reader]
dict_count_children = {}
for child in list_children:
    if child in dict_count_children:
        dict_count_children[child] += 1
    else:
        dict_count_children[child] = 1
print(dict_count_children)
plt.xlabel("Number of children")
plt.ylabel("Number of people")
plt.bar(dict_count_children.keys(),dict_count_children.values())
plt.show()
```

```python
#show how charge depends on age
plt.plot(list_age,list_charges,'bo')
plt.xlabel("Age")
plt.ylabel("Charge")
plt.show()
```

```python
#show how charge depends on bmi
plt.plot(list_bmi,list_charges,'bo')
plt.xlabel("BMI")
plt.ylabel("Charge")
plt.show()
```

```python
#show how charge depends on number of children
plt.plot(list_children,list_charges,'bo')
plt.xlabel("Number of children")
plt.ylabel("Charge")
plt.show()
```

```python
#show how charge depends on regions
plt.plot(list_region,list_charges,'bo')
plt.xlabel("Regions")
plt.ylabel("Charge")
plt.show()
```

```python
#show how charge depends on smoke
plt.plot(list_smoker,list_charges,'bo')
plt.xlabel("Smoke")
plt.ylabel("Charge")
plt.show()
```

After analyzing the graphs and data, we can confidently say that age, smoking and weight significantly affect the cost of health insurance. The region of residence also affects the cost.
